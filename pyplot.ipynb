1. Given the text below, create a bar plot (or a horizontal bar plot) that shows the frequency distribution of characters in the string. 
A dictionary has been created for you containing the characters and the frequencies of the characters.
Make sure your bar plot has the bars labelled correctly (with the character it represents the frequency of).


## Question 1 Solution

```python
from collections import Counter
import matplotlib.pyplot as plt

# Provided text
TEXT = ("At three o’clock precisely I was at Baker Street, but Holmes had not yet returned. "
        "The landlady informed me that he had left the house shortly after eight o’clock in the morning. "
        "I sat down beside the fire, however, with the intention of awaiting him, however long he might be. "
        "I was already deeply interested in his inquiry, for, though it was surrounded by none of the grim and strange features "
        "which were associated with the two crimes which I have already recorded, still, the nature of the case and the exalted "
        "station of his client gave it a character of its own. Indeed, apart from the nature of the investigation which my friend "
        "had on hand, there was something in his masterly grasp of a situation, and his keen, incisive reasoning, which made it a "
        "pleasure to me to study his system of work, and to follow the quick, subtle methods by which he disentangled the most "
        "inextricable mysteries. So accustomed was I to his invariable success that the very possibility of his failing had ceased "
        "to enter into my head.").lower()

# Count character frequencies
letter_freq = dict(Counter(TEXT))

# Sort the dictionary by frequency
sorted_freq = dict(sorted(letter_freq.items(), key=lambda item: item[1], reverse=True))

# Create the horizontal bar plot
plt.figure(figsize=(12, 10))
plt.barh(list(sorted_freq.keys()), list(sorted_freq.values()), color='tan')
plt.xlabel('Frequency')
plt.ylabel('Character')
plt.title('Character Frequency Distribution in Text')
plt.gca().invert_yaxis()  # Highest frequency on top
plt.tight_layout()
plt.show()

```

2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

## Question 2 Solution

```python
import matplotlib.pyplot as plt
import random

# Generate a sequence of at least 5 random integers
random_numbers = [random.randint(1, 100) for _ in range(7)]
indices = list(range(len(random_numbers)))

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Scatter plot
ax1.scatter(indices, random_numbers, color='blue')
ax1.set_title('Scatter Plot of Random Numbers')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value')

# Line plot
ax2.plot(indices, random_numbers, marker='o', color='green')
ax2.set_title('Line Plot of Random Numbers')
ax2.set_xlabel('Index')
ax2.set_ylabel('Value')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()





```

3. Do the same question as above, but change the syle and color of both the scatter and the plot.

## Question 3 Solution

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a sequence of 7 random integers
np.random.seed(0)
data = np.random.randint(1, 100, size=7)
indices = np.arange(len(data))

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Scatter plot with hexagon markers in dark blue
ax1.scatter(indices, data, marker='h', color='darkblue')
ax1.set_title("Scatter Plot of Random Numbers", fontweight='bold')
ax1.set_xlabel("Index", fontweight='bold')
ax1.set_ylabel("Value", fontweight='bold')

# Line plot with hexagon markers and line in dark red
ax2.plot(indices, data, marker='h', color='darkred', markerfacecolor='darkred')
ax2.set_title("Line Plot of Random Numbers", fontweight='bold')
ax2.set_xlabel("Index", fontweight='bold')
ax2.set_ylabel("Value", fontweight='bold')

plt.tight_layout()
plt.show()

```

You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time.  Assume that `insertion_sort_times[i]` is how long it took to sort a list of length `sizes[i]`.  Add a legend to your plot showing which is insertion sort and which is merge sort.  Use the variable names to determine which data is which. 

## Question 4 Solution

```python
import matplotlib.pyplot as plt

# Provided data
sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(sizes, insertion_sort_times, marker='o', label='Insertion Sort')
plt.plot(sizes, merge_sort_times, marker='s', label='Merge Sort')

# Add labels and title
plt.xlabel('List Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Sorting Algorithms')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()

```
